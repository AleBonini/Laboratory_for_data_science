# Match

```python
match = pd.read_csv('match.csv')
```

```python
match
```

```python
match.shape
```

```python
match.isnull().sum()
```

```python
match = match[["tourney_id",
"match_id",
"winner_id",
"loser_id",
"score",
"best_of",
"round",
              "winner_rank",
"winner_rank_points",
"loser_rank",
"loser_rank_points"]]
```

```python
for column in match:
    print (column)
```

```python
match.isnull().sum()
```

```python
for col in match:
    if match[col].dtypes == 'object':
        match[col] = match[col].fillna(match[col].mode()[0])
    else:
        match[col] = match[col].fillna(match[col].median())
match.isnull().sum()
```

```python
match.isnull().sum()
```

```python

match.dtypes
```

```python
match.duplicated().sum()
```

```python
match.drop_duplicates(subset ="match_id", inplace = True)
```

```python
match.to_csv('match_cleaned.csv',index = False)
```
