```python
with open('C:\\Users\\bonin\\lds_project\\date.csv', 'w', newline="") as output:
    with open('C:\\Users\\bonin\\lds_project\\tmp_date.csv', 'r') as date_file:
        
        writer = writer(output)
        writer.writerow(['date_id', 'year', 'month', 'day','quarter'])
        next(date_file)        
        reader = reader(date_file)
        
        for row in reader:

            year = int(row[0][:4])
            month = int(row[0][4:6])
            day = int(row[0][6:8])
            quarter = month//4 + 1

            row.append(year)
            row.append(month)
            row.append(day)
            row.append(quarter)
            
            writer.writerow(row)
```

# Player

```python
female = []
male = []

with open("C:\\Users\\bonin\\lds_project\\female_players.csv", "r") as female_file:
    with open("C:\\Users\\bonin\\lds_project\\male_players.csv", "r") as male_file:
        
        for line in female_file:
            line = line.replace(","," ").strip()
            female.append(line)
            
        for line in male_file:
            line = line.replace(","," ").strip()
            male.append(line)
            
male.pop(0)
female.pop(0)
```

```python
print(male[:5])
print(female[:5])
```

```python
players = {}
with open("C:\\Users\\bonin\\lds_project\\tennis.csv") as tennis_file:
    for i in tennis_file:
        line = i.strip().split(",")
        if line[9] not in players:
            players[line[9]] = [line[7]]
        if line[16] not in players:
            players[line[16]] = [line[14]]
players.pop("winner_name")
players.pop("loser_name")
players
```

```python
male = set(male)
female = set(female)

finale = {}

for i in players:
    if i in female:
        players[i].append("F")
    elif i in male:
        players[i].append("M")

players
```

```python
for i in players:
    if len(players[i]) == 1:
        players[i].append("Null")
```
